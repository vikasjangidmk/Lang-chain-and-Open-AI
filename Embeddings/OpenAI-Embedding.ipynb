# ** Embeddings **
Embeddings are vector representations of text, words, or documents that capture their semantic meaning in a high-dimensional space. These representations are learned from large datasets and help models understand relationships and similarities between different pieces of text.

In LangChain, embeddings can be used for various purposes, such as document retrieval, summarization, and question answering.

---

## ** Common Embedding Models**

LangChain supports a variety of embedding models, including both pre-trained models from OpenAI and other popular language models. Some common models include:

- **OpenAI Embeddings**: Used for transforming text into vectors, such as `text-embedding-ada-002`.
- **Sentence Transformers**: A popular family of models for sentence-level embeddings.
- **Hugging Face Transformers**: A library with various pre-trained models for generating embeddings.

```python
import os
from dotenv import load_dotenv
load_dotenv()   # Load all the env variables
```

```python
os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY")
```

```python
from langchain_openai import OpenAIEmbeddings

embeddings = OpenAIEmbeddings(model="text-embedding-3-large")
```

```python
embeddings
```

```python
text = "This is a tutorial on OpenAI embedding"
query_result = embeddings.embed_query(text)
query_result
```

```python
embeddings_1024 = OpenAIEmbeddings(model="text-embedding-3-large", dimensions=1024)
```

```python
embeddings_1024
```

```python
text = "This is a tutorial on OpenAI embedding"
query_result = embeddings_1024.embed_query(text)
len(query_result)
```

```python
query_result
```

```python
from langchain_community.document_loaders import TextLoader

loader = TextLoader('speech.txt')
docs = loader.load()
docs
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=50)
final_documents = text_splitter.split_documents(docs)
final_documents
```

## Vector Embedding and Vector Store

```python
from langchain_community.vectorstores import Chroma

db = Chroma.from_documents(final_documents, embeddings_1024)
db
```

```python
## Retrieve the results from query vectorstore db

query = "It will be all the easier for us to conduct ourselves as belligerents"
retrieved_result = db.similarity_search(query)
print(retrieved_result)
```
