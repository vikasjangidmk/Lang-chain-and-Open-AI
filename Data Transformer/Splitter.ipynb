# **Text Splitter**

## **What is a Text Splitter?**
A Text Splitter is a utility used to divide large blocks of text into smaller, manageable chunks. This is especially useful in applications where processing, analyzing, or storing large documents requires breaking them into parts, such as paragraphs, sentences, or tokenized chunks.

---

## **Types of Text Splitters**
1. **Character-Based Splitters**  
   Split text based on a fixed number of characters.  
   - **Example**: Breaking a long string into chunks of 100 characters each.

2. **Sentence-Based Splitters**  
   Split text at sentence boundaries.  
   - **Example**: Dividing a document into individual sentences.

3. **Paragraph-Based Splitters**  
   Divide text into paragraphs based on newline or spacing delimiters.  
   - **Example**: Splitting a report into its paragraphs.

4. **Token-Based Splitters**  
   Split text into chunks based on a specified number of tokens, often used in language models.  
   - **Example**: Dividing a text into 500-token chunks.

---

## **Use Cases for Text Splitters**
- **Preprocessing for NLP Models**: Breaking text into manageable pieces for tokenization or embeddings.
- **Chunking for Summarization**: Preparing text for extractive or abstractive summarization tasks.
- **Segmented Data Storage**: Storing text in smaller units for efficient querying and retrieval.
- **Error Handling**: Avoiding input size limits for APIs or models like OpenAI's GPT.

---

## **Steps for Using a Text Splitter**
1. **Initialize the Splitter**  
   Define the type of splitter (e.g., character, sentence, or token-based).
2. **Provide the Text**  
   Input the text to be split into chunks.
3. **Configure Parameters**  
   Set chunk size, overlap size, or other relevant parameters.
4. **Split Text**  
   Use the splitter to divide the text into chunks.

---

## **Example: Using LangChain’s Text Splitter**

### **Character-Based Text Splitter**
```python
from langchain.text_splitter import CharacterTextSplitter

# Initialize the text splitter
splitter = CharacterTextSplitter(
    chunk_size=100,  # Maximum size of each chunk
    chunk_overlap=10  # Number of overlapping characters
)

# Sample text
text = "This is a sample text. It will be split into smaller chunks for processing."

# Split the text
chunks = splitter.split_text(text)

# Print the chunks
for chunk in chunks:
    print(chunk)

# **Recursively Split Text by Characters**

## **What is Recursive Text Splitting?**
Recursive text splitting is the process of breaking a large text into smaller chunks iteratively or hierarchically. If a chunk still exceeds the desired size after the initial split, it is further split recursively until all chunks meet the size requirements.

This method is useful when processing large text documents to ensure compatibility with systems or models that have input size limitations, while preserving as much context as possible.

---

## **Key Steps in Recursive Text Splitting**

1. **Initial Split**  
   Divide the text into chunks based on a primary delimiter, such as paragraphs or sections.  

2. **Size Check**  
   Validate whether each chunk meets the desired size criteria (e.g., maximum character count).  

3. **Recursive Splitting**  
   For chunks that exceed the size limit, split them further using a secondary delimiter, such as sentences or character counts.  

4. **Final Check**  
   Ensure all resulting chunks are within the specified size limit.

---

## **Why Use Recursive Splitting?**
- **Context Preservation**: Maintains hierarchical structure by splitting at logical points (e.g., paragraphs before sentences).  
- **Efficiency**: Reduces the need for manual preprocessing of large documents.  
- **Scalability**: Ensures compatibility with systems that have strict input size limits.  

## **How It Works**
1. Define Separators
Provide a list of delimiters in order of preference, such as paragraphs (\n\n), sentences (\n), and spaces ( ).

2. Set Chunk Parameters
Specify chunk_size (maximum size of each chunk) and chunk_overlap (number of characters overlapping between chunks).

3. Recursive Splitting
The splitter uses the separators iteratively. If a chunk still exceeds the limit, the next delimiter in the list is used for further splitting.

## **Advantages of Recursive Splitting**
- Flexibility: Adapts to varying text structures and sizes.
- Granularity: Ensures text is split logically while preserving important contextual boundaries.
- Customizability: Users can define their own delimiters and chunk parameters.

## **Challenges and Best Practices**
- Separator Selection
Choose meaningful delimiters to avoid breaking sentences or words unnecessarily.

- Overlap Setting
Use overlaps to maintain context between adjacent chunks for better downstream processing.

- Performance Considerations
For very large texts, optimize memory usage to handle splitting efficiently.

---

## **Example: Recursive Splitting Using LangChain**

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

# Initialize the recursive character splitter
splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,  # Maximum size of each chunk
    chunk_overlap=50,  # Number of overlapping characters between chunks
    separators=["\n\n", "\n", " "]  # Delimiters for splitting (paragraph, sentence, word)
)

# Sample text
text = """This is a large block of text. It contains multiple paragraphs, sentences, and words.
The goal is to split this text into smaller chunks recursively, maintaining context.
For instance, we may split by paragraphs first, then sentences, and finally words."""

# Recursively split the text
chunks = splitter.split_text(text)

# Print the resulting chunks
for idx, chunk in enumerate(chunks):
    print(f"Chunk {idx + 1}:")
    print(chunk)
    print("-" * 40)

# **Split by HTML header**

HTMLHeaderTextSplitter is a "structure-aware" chunker that splits text at the HTML element level and adds metadata for each header "relevant" to any given chunk. It can return chunks element by element or combine elements with the same metadata, with the objective of (a) keeping related text grouped (more or less) semantically and (b) preserving context-rich information encoded in document structures. It can be used with other text splitters as part of a chunking pipeline.

````python
from langchain_text_splitters import HTMLHeaderTextSplitter

html_string = """
<!DOCTYPE html>
<html>
<body>
    <div>
        <h1>Foo</h1>
        <p>Some intro text about Foo.</p>
        <div>
            <h2>Bar main section</h2>
            <p>Some intro text about Bar.</p>
            <h3>Bar subsection 1</h3>
            <p>Some text about the first subtopic of Bar.</p>
            <h3>Bar subsection 2</h3>
            <p>Some text about the second subtopic of Bar.</p>
        </div>
        <div>
            <h2>Baz</h2>
            <p>Some text about Baz</p>
        </div>
        <br>
        <p>Some concluding text about Foo</p>
    </div>
</body>
</html>
"""

headers_to_split_on=[
    ("h1","Header 1"),
    ("h2","Header 2"),
    ("h3","Header 3")
]

html_splitter=HTMLHeaderTextSplitter(headers_to_split_on)
html_header_splits=html_splitter.split_text(html_string)
html_header_splits
````

````python
url = "https://plato.stanford.edu/entries/goedel/"

headers_to_split_on = [
    ("h1", "Header 1"),
    ("h2", "Header 2"),
    ("h3", "Header 3"),
    ("h4", "Header 4"),
]
html_splitter = HTMLHeaderTextSplitter(headers_to_split_on)
html_header_splits = html_splitter.split_text_from_url(url)
html_header_splits
````

#### How to split JSON data
This json splitter splits json data while allowing control over chunk sizes. It traverses json data depth first and builds smaller json chunks. It attempts to keep nested json objects whole but will split them if needed to keep chunks between a min_chunk_size and the max_chunk_size.

If the value is not a nested json, but rather a very large string the string will not be split. If you need a hard cap on the chunk size consider composing this with a Recursive Text splitter on those chunks. There is an optional pre-processing step to split lists, by first converting them to json (dict) and then splitting them as such.

- How the text is split: json value.
- How the chunk size is measured: by number of characters.


````python
import json
import requests

json_data=requests.get("https://api.smith.langchain.com/openapi.json").json()
````

````python
json_data
````

````python
from langchain_text_splitters import RecursiveJsonSplitter
json_splitter=RecursiveJsonSplitter(max_chunk_size=300)
json_chunks=json_splitter.split_json(json_data)
````

````python
json_chunks
````

````python
for chunk in json_chunks[:3]:
    print(chunk)
````

````python
## The splitter can also output documents
docs=json_splitter.create_documents(texts=[json_data])
for doc in docs[:3]:
    print(doc)
````

````python
texts=json_splitter.split_text(json_data)
print(texts[0])
print(texts[1])
````
